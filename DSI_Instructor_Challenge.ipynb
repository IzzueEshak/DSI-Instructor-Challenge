````python
def bootsrap_sample(df,sample:int):
    df = DataFrame(df) # ensure input is a pandas dataframe 
    size = sample/len(df)
    return df.sample(frac = size,replace = True)

bootsrap_sample(bcancer,5)
````

### Data Pre-Processing

````python
#separating target variable (diagnosis) from initial dataframe 
target_v = bcancer['diagnosis']
input_v = bcancer_final[['fractal_dimension_mean','concave_points_sd_error_log','perimeter_sd_error_log','radius_mean_log','perimeter_mean_log']]
````

````python
input_v.describe()
````

````python
from sklearn.model_selection import train_test_split, cross_val_score,cross_val_predict,GridSearchCV

#Spliting train and test datasets on a 70 percent to train and 30 percent to test ratio
X_train, X_test, y_train, y_test = train_test_split(input_v,target_v,test_size=0.3,random_state = 42)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

scaledX_train = sc.fit_transform(X_train)
scaledX_test = sc.fit_transform(X_test)
````

## Model building
#### SVM Model and Random Forest Model will be built for this data
### SVM
#### Advantages

1. Easily interpretable and performs faster compared to more advanced models
2. Uses less memory because they use a subset of training points in the decision phase
3. SVM works well with a clear margin of separation and with high dimensional space.

#### Disadvantages

1. Sensitive to the types of kernels
2. Works poorly with overlapping classes

### Random Forest
#### Advantages

1. It reduces overfitting in decision trees and helps to improve the accuracy
2. It is flexible to both classification and regression problems
3. It works well with both categorical and continuous values
4. It automates missing values present in the data if any

#### Disadvantages

1. It is not easily interpretable
2. It requires much computational power as well as resources as it builds numerous trees to combine their outputs.
3. It also requires much time for training as it combines a lot of decision trees to determine the class.

### SVM Model

````python
#Import svm model
from sklearn import svm
#Create a svm Classifier
clf_SVM = svm.SVC(kernel='linear') # Using Linear Kernel

#Grid search for SVM
Cs = [0.001, 0.01, 0.1, 1, 10,100,1000]
gammas = [0.001, 0.01, 0.1, 1,10,100]
optimum_grid_search=GridSearchCV(clf_SVM, {'C': Cs, 'gamma' : gammas})
optimum_grid_search.fit(scaledX_train,y_train)
optimum_grid_search.best_params_
````

````python
from sklearn import metrics

#Creating SVM classifier with given C and Gamma values
clf_SVM = svm.SVC(kernel='linear',C=10, gamma=0.001) 

#Train the model using the training sets
clf_SVM.fit(scaledX_train, y_train)

#Predict the response for test dataset
y_pred = clf_SVM.predict(scaledX_test)

print("SVM Evaluation Matrix")
# Model Accuracy: how often is the classifier prediction correct?
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))

#from sklearn.metrics import confusion_matrix
print("Confusion matrix:", metrics.confusion_matrix(y_test, y_pred))

#ROC AUC Score
print("ROC-AUC Score:", metrics.roc_auc_score(y_test, y_pred))
````

For SVM, Based on the confusion matrix it is identified that the True Negative value which is actually no and predicted no is 105 and the True Positive value which is actually yes and predicted yes is 59. The model accuracy is 95.9 % and has a ROC-AUC score of 95.4%. Grid search method was used to reduce the effect of overfitting.

#### Identifying important features in SVM
Important features in SVM model can be identified by using the coef property

````python

feature_names = X_train.columns
SVM_features = DataFrame({'Column': feature_names, 'coef_weights':np.abs(clf_SVM.coef_[0])}).sort_values(ascending = False, by = 'coef_weights')
SVM_features.reset_index(drop=True, inplace=True)

print("Top 3 important features identified by SVM are:")
SVM_features['Column'].head(3)
````

### Random Forest Model

````python

from sklearn.ensemble import RandomForestClassifier
clf_randomforest = RandomForestClassifier(random_state=42)

#Grid Search for Random Forest
param_grid = {
    'bootstrap': [True],
    'max_depth': [2,3,4,6],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [4,8,12],
    'n_estimators': [10,50,100,150,200]
}

optimum_grid_search = GridSearchCV(estimator = clf_randomforest, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
optimum_grid_search.fit(scaledX_train, y_train)
optimum_grid_search.best_params_
````

````python
clf_rf_best_search = optimum_grid_search.best_estimator_

#Predict the response for test dataset based on best estimators from grid search
rf_y_pred = clf_rf_best_search.predict(scaledX_test)

print("Random Forest Evaluation Matrix")
# Model Accuracy: how often is the classifier prediction correct?
print("Accuracy:",metrics.accuracy_score(y_test, rf_y_pred))

#from sklearn.metrics import confusion_matrix
print("Confusion matrix:",metrics.confusion_matrix(y_test, rf_y_pred))

#ROC AUC Score
print("ROC-AUC Score:",metrics.roc_auc_score(y_test, rf_y_pred))
````

For Random Forest, based on the confusion matrix it is identified that the True Negative value which is actually no and predicted no is 105 and the True Positive value which is actually yes and predicted yes is 57. The model accuracy is 94.7 % and has a ROC-AUC score of 93.8%. Grid search method was used to reduce the effect of overfitting here as well

#### Identifying important features in Random Forest

Important features in SVM model can be identifyed by using the feature_importances property

````python
rf_features = DataFrame({'Column': feature_names, 'coef_weights':np.abs(clf_rf_best_search.feature_importances_)}).sort_values(ascending = False, by = 'coef_weights')
rf_features.reset_index(drop=True, inplace=True)

print("Top 3 important features identified by Random Forest are:")
rf_features['Column'].head(3)
````
